```python
def imput_edu(cols):
    education=cols[0]
    department = cols[1]
    if pd.isnull(education):
        if department=="R&D" or "Sales & Marketing" or "Operations" or "HR":
            return "Master's&above"
        elif department=="Analytics" or "Technology":
            return "Bachelor's"
        else:
        
            return "secondary"
        
    else:
        return education
        
    
```

```python
hr['education']=hr[['education','department']].apply(imput_edu,axis=1)
```

```python
def imput_rat(cols):
    previous_year_rating=cols[0]
    department = cols[1]
    if pd.isnull(previous_year_rating):
        if department=="R&D" or "Sales & Marketing" or "Operations" or "HR":
            return 5
        elif department=="Analytics" or "Technology":
            return 4
        else:
        
            return 3
        
    else:
        return previous_year_rating
        
```

```python
hr_feat=pd.DataFrame(scaled_feat,columns=hr.columns[:-1])
```

```python
x=hr_feat
z=y
```

```python
 X_train, X_test, y_train, y_test = train_test_split(x, z, test_size=0.2, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(X_train,y_train)
```

```python
predict=knn.predict(X_test)
```

```python
predict
```

```python
from sklearn.metrics import classification_report,confusion_matrix 
```

```python
from sklearn.metrics import f1_score
```

```python
print(classification_report(predict,y_test))
```

```python
f1_score(y_test,predict, average='micro')
```

```python
print(confusion_matrix(predict,y_test))
```

```python
err_rate=[]
for i in range(1,40):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pre=knn.predict(X_test)
    err_rate.append(np.mean(pre!=y_test))

```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),err_rate,color='blue',linestyle='dashed',marker='o',markerfacecolor='red',markersize=10)
plt.title('error rate vs k value')
plt.xlabel('k')
plt.ylabel('err_rate')
```
